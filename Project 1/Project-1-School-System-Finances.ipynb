```python
plt.figure(1)
plt.figure(figsize=(15,10))
plt.scatter(revenue, spending)
myOLS_points = sm.OLS(revenue,revenue).fit()
plt.plot(revenue, myOLS_points.predict(revenue), color = 'red')
plt.title("Spending vs Revenue")
plt.xlabel("Revenue")
plt.xlim([0, 3*10e5])
plt.ylim([0, 3*10e5])
plt.ylabel("Spending")
plt.show()
plt.close()
```
